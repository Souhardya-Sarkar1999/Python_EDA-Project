## Data Preprocessing

```python
# --- 1. Handle missing values ---
# Drop columns with >50% missing values
df = df.loc[:, df.isnull().mean() < 0.5]
```

```python
# Drop rows missing critical information
df = df.dropna(subset=["Title", "Start Date", "Completion Date"])
```

```python
# --- 2. Convert dates ---

from datetime import datetime 

date_cols = ["Start Date", "Primary Completion Date", "Completion Date", "First Posted", "Last Update Posted"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# --- 3. Create new feature: Study Duration ---
df["Study Duration (days)"] = (df["Completion Date"] - df["Start Date"]).dt.days
```

```python
# --- 4. Normalize categorical text ---
df["Conditions"] = df["Conditions"].str.lower().str.strip()
df["Status"] = df["Status"].str.title()
df["Gender"] = df["Gender"].fillna("All")
```

```python
# --- 5. Clean text fields ---
df["Title"] = df["Title"].str.strip().str.replace("\s+", " ", regex=True)
```

```python
# --- 6. Convert numeric fields ---
df["Enrollment"] = pd.to_numeric(df["Enrollment"], errors='coerce')
```

```python
# --- 7. Remove duplicates ---
df = df.drop_duplicates(subset="NCT Number")
```

```python
# --- 8. Optional: fill or label remaining missing values for safe analysis ---
df.fillna("Not Specified", inplace=True)
```

# üìä EDA with Visualizations

## A. üü¢ Study Status Distribution

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Status', data=df, order=df['Status'].value_counts().index, palette='viridis')
plt.title('Distribution of Study Status')
plt.xlabel('Count')
plt.ylabel('Status')
plt.tight_layout()
plt.show()
```

### Outcome -

#### * This visualization illustrates the distribution of different study statuses (like Completed, Recruiting, etc.) in the COVID-19 clinical trials dataset.
#### * Most clinical trials are in 'Completed' or 'Recruiting' status, indicating active and concluded research phases.

## B. üìà Enrollment Distribution (log scale)

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Enrollment'], bins=50, kde=True, log_scale=(False, True), color='teal')
plt.title('Enrollment Size Distribution (Log Scale)')
plt.xlabel('Enrollment')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Outcome - 

#### * This plot represents the distribution of participant enrollments in the trials, using a logarithmic scale to handle large variability.
#### * Enrollment numbers vary significantly; some studies involve thousands of participants, while many are small-scale trials.

## C. üìÖ Number of Trials Over Time

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

df['Start Year'] = df['Start Date'].dt.year

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Start Year', order=sorted(df['Start Year'].dropna().unique()), palette='coolwarm')
plt.title('Number of Trials Started Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Trials')
plt.tight_layout()
plt.show()
```

### Outcome - 

#### There was a significant spike in the number of trials initiated in 2020, peaking during the early months of the pandemic.

## D. üîç Top 15 Conditions Studied

```python
from collections import Counter

# Split and normalize condition strings
all_conditions = df['Conditions'].dropna().str.split('|').sum()
top_conditions = Counter(all_conditions).most_common(15)

# Prepare for plotting
conditions, counts = zip(*top_conditions)

plt.figure(figsize=(12,6))
sns.barplot(x=list(counts), y=list(conditions), palette='mako')
plt.title('Top 15 Most Commonly Studied Conditions')
plt.xlabel('Number of Studies')
plt.ylabel('Condition')
plt.tight_layout()
plt.show()
```

### Outcome -

#### Commonly studied conditions include Covid-19, sars -cov2, Anxity and other COVID-related complications.

## E. ‚òÅÔ∏è Word Cloud for Study Titles

```python
text = ' '.join(df['Title'].dropna().values)
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='inferno').generate(text)

plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Clinical Trial Titles')
plt.tight_layout()
plt.show()
```

### Outcome -

## F. üåç Trials by Region (Based on Location Column)

```python
# Extract countries from Location (simplified approach)
df['Country'] = df['Locations'].str.extract(r'([A-Za-z ]+)$')  # last part assumed as country

plt.figure(figsize=(12,6))
top_countries = df['Country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Blues_r')
plt.title('Top 10 Countries Hosting Trials')
plt.xlabel('Number of Trials')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Outcome - 

#### The majority of trials were conducted in the United States, followed by European and Asian countries.

## G. üìÖ Timeline of Trials ‚Äì Monthly Trend

```python
# Convert Start Date to monthly period
df['Start Month'] = df['Start Date'].dt.to_period('M').astype(str)

# Monthly count of trial starts
monthly_counts = df['Start Month'].value_counts().sort_index()
months = monthly_counts.index
counts = monthly_counts.values

# Plot setup
plt.figure(figsize=(14, 6))
sns.lineplot(x=months, y=counts, marker='o', color='darkgreen')

# Clean x-axis ticks
interval = 3  # Change to 4 or 6 for fewer labels
plt.xticks(ticks=range(0, len(months), interval), labels=months[::interval], rotation=45)

# Titles and labels
plt.title('Monthly Trend of Trial Starts')
plt.xlabel('Month')
plt.ylabel('Number of Trials Started')
plt.tight_layout()
plt.show()
```

### Outcome - 

#### * A monthly breakdown of trial initiation trends, reflecting the intensity and timing of global research efforts. 
#### * A monthly surge in trials is noticeable around April to July 2020, coinciding with the global escalation of cases.

## J.üßç‚Äç‚ôÇÔ∏è Age Group Distribution

```python
print(df['Age'].unique())
```

```python
# Function to classify age groups
def classify_age_group(age_str):
    age_str = str(age_str).lower()
    if "child" in age_str:
        return "Child"
    elif "older adult" in age_str or "senior" in age_str or "65" in age_str:
        return "Older Adult"
    elif "adult" in age_str:
        return "Adult"
    elif "18" in age_str and "65" in age_str:
        return "Adult"
    elif "birth" in age_str or "0" in age_str:
        return "Child"
    else:
        return "Unknown"

# Apply classification
df['Age Category'] = df['Age'].apply(classify_age_group)

# Count the number of studies in each age category
age_category_counts = df['Age Category'].value_counts()

# Plotting the distribution
plt.figure(figsize=(8, 5))
age_category_counts.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Distribution of Age Categories in COVID Clinical Trials')
plt.xlabel('Age Category')
plt.ylabel('Number of Studies')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Outcome -

#### * Shows how clinical trial participants were distributed across various age groups.
#### * Trials predominantly targeted adults and older adults, with fewer studies focused exclusively on pediatric populations.
